```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Projects\E-Commerce-Customer-Segmentation-Purchase-Prediction\Data\data.csv", encoding='latin-1')
```

```python
rfm = pd.read_csv(r"D:\Projects\E-Commerce-Customer-Segmentation-Purchase-Prediction\Data\rfm_final.csv", encoding='latin-1')
```

```python
df.isnull().sum()
```

```python
# removing null values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Convert types
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# Create TotalAmount
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

Merging the dataset


```python
merged_df = df.merge(rfm, on="CustomerID", how="inner")
```

```python
merged_df
```

```python
final_df = merged_df[[
    'CustomerID',
    'InvoiceNo',
    'InvoiceDate',
    'Country',
    'Quantity',
    'UnitPrice',
    'TotalAmount',
    'Recency',
    'Frequency',
    'Monetary',
    'Cluster',
    'Segment'
]]
```

```python
final_df
```

```python
final_df.to_csv(r"..\Data\cleaned_data.csv", index=False)
```
